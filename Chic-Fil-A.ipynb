Import sqlite3 for use of the database

```python
#setup sqlite3
import sqlite3
```

```python
#connect to ChicFilA Database
db = sqlite3.connect('ChicFilA.db')
cursor = db.cursor()
```

Here I drop and then redefine a table Menu. This keeps the original Rewards table intact while modifying the Menu table to calculate values based on user imputs

*Future idea: use api to get local tax rate based on ip location*

```python
sql11 = """
DROP TABLE IF EXISTS Working
"""

cursor.execute(sql11)
for r in cursor:
    print(r)
```

```python
sql12 = """
CREATE TABLE IF NOT EXISTS Menu(
    "Name" TEXT,
    "Meal" TEXT,
    "Points" INTEGER,
    "Reward Rank" TEXT,
    "Price w/o Tax" REAL
); 
"""
cursor.execute(sql12)
for r in cursor:
    print(r)
```

```python
sql13 = """
INSERT INTO Menu SELECT * FROM Rewards;
"""
cursor.execute(sql13)
for r in cursor:
    print(r)
```

Add two more columns- Price w/ Tax, then calculate our price per point

```python
sql21 = """
ALTER TABLE Menu
ADD "Price w/ Tax" REAL
"""
cursor.execute(sql21)
for r in cursor:
    print(r)
```

```python
localTaxRate = .075
sql22 = """
UPDATE Menu
SET "Price w/ Tax" = Round(("Price w/o Tax" + ("Price w/o Tax"*"""localTaxRate""")), 2);
"""
cursor.execute(sql22)
for r in cursor:
    print(r)
```

```python
sql21 = """
ALTER TABLE Menu
ADD "Price per Point" REAL
"""
cursor.execute(sql21)
for r in cursor:
    print(r)
```

```python
sql24 = """
UPDATE Menu
SET "Price per Point" = ROUND((("Price w/ Tax"/"Points")*100), 6);
"""
cursor.execute(sql24)
for r in cursor:
    print(r)
```

```python
sql222 = """
SELECT * FROM Menu
"""
cursor.execute(sql222)
for r in cursor:
    print(r)
```

Now the we have calculated our price per point using the local tax rate, we can start to run some selections to make use of our data. The first selection I want to run is what is the most efficient use of my CFA points, i.e. the lowest price per point

```python
sql31 = """
SELECT "Name" FROM Menu
WHERE "Price per Point" = (
    SELECT MAX("Price per Point")
    FROM Menu
);
"""
cursor.execute(sql31)
for r in cursor:
    print(r)
```

Say I am going for chic fil a breakfast one morning, and I don't have a ton of points to use. I am going to query the three breakfast items that cost the least

```python
sql32 = """
SELECT "Name" FROM Menu M
WHERE "Price per Point" = (
    SELECT MAX("Price per Point")
    FROM Menu M
    WHERE M.Meal = "Breakfast"
);
"""
cursor.execute(sql32)
for r in cursor:
    print(r)
```
