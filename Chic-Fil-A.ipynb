## Intro

Welcome to Chic-fil-a Database! This is a project with the goal to find the most price effective use of points on the Chic-Fil-A one app. The worksheet was written in Python as a Jupyter Notebook, running queries from an SQL database with the SQLite database engine.

## Schema

Rewards (
<br>
    &emsp;"Name" TEXT
    <br>
    &emsp;"Meal" TEXT
    <br>
    &emsp;"Points" INTEGER
    <br>
    &emsp;"Reward Rank" TEXT
    <br>
    &emsp;"Price" REAL
    <br>
    &emsp;"Price per Point" REAL
    <br>
)
<br>

Import sqlite3 for use of the database, and connect to the database with the Chic-Fil-A One information

```python
# Setup sqlite3 and pandas
import sqlite3
import pandas as pd
```

```python
# Connect to ChicFilA Database
db = sqlite3.connect('ChicFilA.db')
cursor = db.cursor()
```

## Scenario 1

Let's find the information we came for! What are the top five menu items with the best value?

```python
sql1 = """
SELECT * FROM Rewards R
ORDER BY "Price per Point" DESC
LIMIT 5
"""
cursor.execute(sql1)
df = pd.DataFrame()
for r in cursor:
    d = {
        "Name": r[0],
        "Meal": r[1],
        "Points": r[2],
        "Reward Rank": r[3],
        "Price": r[4],
        "Price Per Point": r[5]
    }
    df = df.append(pd.Series(d), ignore_index=True)
display(df)
```

## Scenario 2

Let's say we only have 500 points. What are the best value items under 500 points?

```python
sql2 = """
SELECT * FROM Rewards R
WHERE "Points" < 500
ORDER BY "Price per Point" DESC
LIMIT 3
"""
cursor.execute(sql2)
df = pd.DataFrame()
for r in cursor:
    d = {
        "Name": r[0],
        "Meal": r[1],
        "Points": r[2],
        "Reward Rank": r[3],
        "Price": r[4],
        "Price Per Point": r[5]
    }
    df = df.append(pd.Series(d), ignore_index=True)
display(df)
```

## Scenario 3

I am going for breakfast and want to spend my points wisely. Which items should I avoid?

```python
sql3 = """
SELECT * FROM Rewards R
WHERE "Meal" = "Breakfast"
ORDER BY "Price per Point" ASC
LIMIT 3
"""
cursor.execute(sql3)
df = pd.DataFrame()
for r in cursor:
    d = {
        "Name": r[0],
        "Meal": r[1],
        "Points": r[2],
        "Reward Rank": r[3],
        "Price": r[4],
        "Price Per Point": r[5]
    }
    df = df.append(pd.Series(d), ignore_index=True)
display(df)
```

## Scenario 4

I'm really craving a sandwich. Which kinds have the best point value?

```python
sql4 = """
SELECT * FROM Rewards R
WHERE "Name" LIKE "%sandwich%"
ORDER BY "Price per Point" DESC
LIMIT 3
"""
cursor.execute(sql4)
df = pd.DataFrame()
for r in cursor:
    d = {
        "Name": r[0],
        "Meal": r[1],
        "Points": r[2],
        "Reward Rank": r[3],
        "Price": r[4],
        "Price Per Point": r[5]
    }
    df = df.append(pd.Series(d), ignore_index=True)
display(df)
```
